```python
import re
```

```python
# re.match(pattern, str)

t_str = '이사원, 오대리, 구과장, 육차장, 정부장'

print(re.match('이사원', t_str))
print(re.match('오대리',t_str))
```

```python
# re.search(pattern, str)

print(re.search('이사원', t_str))
print(re.search('오대리',t_str))
```

```python
# re.findall(pattern, str)

print(re.findall('장', t_str))

# re.split

words = re.split(',',t_str)
for w in words:
  if re.search('장',w):
    print(w)
```

```python
# re.sub(target, sub, str)

t_str = re.sub('구', '9', t_str)
print(t_str)
```

```python
# findall

re.findall('0123456789', t_str) # 통 일치

re.findall('[0123456789]', t_str) # 일부 일치 발췌
re.findall('[0-9]', t_str)

for w in words:
  if re.findall('과장|차장|부장', w):
    print(w)
```

```python
import re

# Phone number Cleaning

phone = ['010-1111-2222', '010-12345678', '010 5555 7777', '07009991000', '020933644', '010.2222~5555']

phone_pattern = re.compile(r'010[-=~ \.]?[0-9]{4}[-=~ \.]?[0-9]{4}')

p_revised = list()
# 'i' 변수는 append() 메소드를 사용하면 필요 없습니다.

for p_original in phone:
  if phone_pattern.search(p_original):
    cleaned_p = re.sub('[^0-9]', '', p_original)
    if len(cleaned_p) == 11 and cleaned_p.startswith('010'):
      formatted_p = f"{cleaned_p[:3]}-{cleaned_p[3:7]}-{cleaned_p[7:]}"
      p_revised.append(formatted_p)
      print(formatted_p)
    else:
      p_revised.append(cleaned_p)
      print(cleaned_p)
```
