```python
import pandas as pd
import numpy as np

data = {
    'Samsung': [80000, 81000, np.nan, 80500, np.nan],   # 2일치 누락
    'SK_Hynix': [120000, np.nan, 122000, 121000, 123000], # 1일치 누락
    'Volume': [1000, 2000, 1500, None, 3000]              # None도 NaN으로 변환됨
}

# 2. 날짜 인덱스 만들기
dates = pd.date_range(start='2026-01-20', periods=5)

# 3. 데이터프레임 생성
df = pd.DataFrame(data, index=dates)
print(df)
```

```python
print(df)
print(df.isnull().sum())
df = df.fillna(method="ffill")
if df.isnull().values.any():
  df = df.fillna(method="bfill")
print(df)
```
