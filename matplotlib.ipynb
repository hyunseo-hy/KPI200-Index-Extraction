```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from matplotlib import font_manager
import pandas as pd
```

```python
x = np.arange(100,300,10)
y = np.random.randn(20)
plt.plot(x,y)
```

```python
set(sorted([f.name for f in font_manager.fontManager.ttflist]))
```

```python
font_title = {"family": "Malgun Gothic", "color": "black", "weight": "bold", "size": 20}
font_label = {"family": "Malgun Gothic", "color": "black", "weight": "bold", "size":10}
plt.plot(x,y,color='m', linestyle=":", marker="*", markersize="20", label = "1")
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.plot(x,y+1,color='r', linestyle="-", label = "2")
plt.title("", fontdict = font_title)
plt.xlabel("x", fontdict = font_label)
plt.ylabel("y", fontdict = font_label)
plt.grid(True)
plt.legend(loc="best")
```

```python
'''
   
'''
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False 

fig = plt.figure(figsize=(16,10))

ax1=fig.add_subplot(2,3,1)
ax1.plot(x,y)

ax2=fig.add_subplot(2,3,2)
ax2.hist(y)

ax3=fig.add_subplot(2,3,3)
ax3.scatter(x,y)

ax4=fig.add_subplot(2,3,4)
ax4.plot(x,y,color='r',linestyle='--')
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

fig, axes = plt.subplots(2, 3, figsize=(16, 10))

axes[0, 0].plot(x, y)
axes[0, 1].hist(y)
axes[0, 2].scatter(x, y)
axes[1, 0].plot(x, y, color='r', linestyle='--')

plt.tight_layout()  #    
plt.show()
```
